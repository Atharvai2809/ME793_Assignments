```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### First we convert the given .txt file into a .csv file

```python
df=pd.read_csv('material_data_Alankar.csv')
```

```python
Atomic_No=df['Atomic_Number']
Electronegativity=df['Electronegativity']
```

```python
plt.plot(Atomic_No,Electronegativity)
plt.title('Electronegativity vs Atomic Number Plot')
plt.xlabel('Atomic Number')
plt.ylabel('Electronegativity')
plt.show()
```

### From the above plot we can deduce that the number of peaks are 5. So while clustering them using K-means Algorithm, we will take the number of clusters to be 5

```python

data = list(zip(Atomic_No, Electronegativity))
kmeans = KMeans(n_clusters=5)
kmeans.fit(data)

plt.scatter(Atomic_No, Electronegativity, c=kmeans.labels_)
plt.title('Clustering using K-means')
plt.xlabel('Atomic Number')
plt.ylabel('Electronegativity')
plt.show()
```
